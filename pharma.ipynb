```
import pandas as pd
```

```
df = pd.read_csv('/content/train.csv')
del df['id']
df
```

```
df.describe()
```

```
df['species'].value_counts()
```

-As we can see from above result given dataset in perfectly balanced

```
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
le.fit(df['species'])
encoded_values = le.transform(df['species'])
encoded_values
```

```
le.classes_
```

```
df['species']= encoded_values
df.shape
```

```
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn import svm
```

```
X = df.iloc[:,1:193]
y = df.iloc[:,0]
```

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```
rf_clf = RandomForestClassifier(random_state=0)
rf_clf.fit(X_train, y_train)
```

```
y_pred_rf = rf_clf.predict(X_test)
y_pred_rf
```

```
dt_clf = DecisionTreeClassifier(random_state=0)
dt_clf.fit(X_train, y_train)
```

```
y_pred_dt = dt_clf.predict(X_test)
y_pred_dt
```

```
svm_clf = svm.SVC()
svm_clf.fit(X_train, y_train)
```

```
y_pred_svm = svm_clf.predict(X_test)
y_pred_svm
```

```
nb_clf = GaussianNB()
nb_clf.fit(X_train, y_train)
```

```
y_pred_nb = nb_clf.predict(X_test)
y_pred_nb
```

```
from sklearn.metrics import accuracy_score
print(f'ACCuracy of RandomForest->> {accuracy_score(y_test,y_pred_rf)}')
print(f'ACCuracy of DecisionTree->> {accuracy_score(y_test,y_pred_dt)}')
print(f'ACCuracy of SVM->> {accuracy_score(y_test,y_pred_svm)}')
print(f'ACCuracy of NaiveBaysian->> {accuracy_score(y_test,y_pred_nb)}')
```

-Most accurate model : **RandomForest**

```
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred_rf))
```

-Random Forest is giving highest accuracy
